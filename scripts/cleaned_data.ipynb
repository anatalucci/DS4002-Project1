# Clean Data

```python
import pandas as pd
df = pd.read_csv("most_reviewed_asin.csv") #upload dataset
df.head()
```

```python
df = df[df["verified_purchase"] == True] #removed rows where verified purchase is false 
```

```python
len(df) 
df["verified_purchase"].value_counts() #check to verify rows were remvoed
```

```python
df["asin"].value_counts() #check to make sure all of the rows in the dataset are for the firestick 
```

```python
df.columns
```

```python
df = df.drop(columns=["images", "parent_asin", "user_id", "helpful_vote"]) #drop unnecessary columns
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df['timestamp'] = df['timestamp'].dt.year.astype(str)
```

```python
df.to_csv("firestick_reviews_cleaned.csv", index=False)
```

I dropped the columns images, parent_asin, helpful_vote, and user_id. Images and user_id didn't contribute anything for the purposes of our project. Asin and parent_asin are both product ids, but I kept asin becuase "B00ZV9RDKK" is the product id for the amazon firestick so I dropped any rows with a different asin because it meant it wasn't a firestick. I got rid of parent_asin because it was just a more specific product id and didn't really matter. Helpful vote also didn't really do anything for our purposes because all it tells us is how many people found the review helpful, which doesn't indicate anything about the actual positivity or negativity of the review. Then I also removed all rows that had "False" under verified purchase and also reduced the timestamp column down to just the year.
