```python
import pandas as pd
from collections import Counter

asin_counter = Counter()

chunks = pd.read_json(
    "Electronics.jsonl",
    lines=True,
    chunksize=50_000
)

for chunk in chunks:
    asin_counter.update(chunk['asin'])

top_asin, top_count = asin_counter.most_common(1)[0]

top_asin, top_count
```

```python
rows = []

chunks = pd.read_json(
    "Electronics.jsonl",
    lines=True,
    chunksize=50_000
)

for chunk in chunks:
    rows.append(chunk[chunk['asin'] == top_asin])

df_top = pd.concat(rows, ignore_index=True)

df_top.head()
```

```python
df_top.to_csv("most_reviewed_asin.csv", index=False)
```
